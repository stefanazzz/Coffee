### Various modules loaded

```python
# plotting
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
matplotlib.interactive(True)
import ipywidgets as widgets
from ipywidgets import *
# numeric and signal proc
import numpy as np
from scipy import ndimage,arange
from scipy import signal as sig
from scipy.fftpack import fft, ifft
# image proc
from PIL import Image, ImageFilter, ImageEnhance
from pylab import plot, xlabel, ylabel
# various
from IPython.display import display, Math, Latex
import time
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
# read and reshape a files with 2 columns:
thick1=pd.read_csv('./examplefile',header=None,delim_whitespace=True);
time=np.array(thick1.iloc[:,0]);
thick=np.array(thick1.iloc[:,1]);
dt=time[1]-time[0];
print(dt)
print(thick1.shape)
print(len(thick))
```

```python
# compute two reference curves:
line=np.zeros(len(thick))
diff=np.zeros(len(thick))
for i in range(len(thick)):
    line[i]=1e-9*float(i)
    diff[i]=np.sqrt(1e-10*dt*float(i))
```

```python
# three subplots in same figure:
fig,ax=plt.subplots(3)
ax[0].plot(time,thick);
ax[1].plot(time,line);
ax[2].plot(time,diff);
```

```python
# combined curves in one plot:
fig,ax=plt.subplots(1)
ax.loglog(time,thick);
ax.loglog(time,line,'--')
ax.loglog(time,diff,'--')
ax.set_xlim(1e0,1e2)
ax.set_ylim(1e-7,1e-4)
ax.set_xlabel('time (s)')
ax.set_ylabel('melt thickness (m)');
```
