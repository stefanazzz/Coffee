## Specific libraries for image proc (PIL, cv2)<br> and OCR (pytesseract):

```python
from PIL import Image
import pytesseract
import cv2
```

## Import bitmap image<br>Save to temporary file<br>(and optional preview of bitmap):

```python
image = cv2.imread("example.png") #read using cv2 module
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # convert to gray numbers using cv2
filename = "{}.png".format("temp") # create temp file 
cv2.imwrite(filename, gray) # save as temp file
```

```python
# optional preview of image:
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
plt.imshow(image);
```

## Interpret as text using tesseract module: 

```python
text = pytesseract.image_to_string(Image.open(filename))
print(text)
```
